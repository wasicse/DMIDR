```python
from Bio import SeqIO

def filter_fasta_by_partial_id(fasta_file, id_list_file, output_file):
    # Read the list of IDs from the file and store as prefixes
    with open(id_list_file) as f:
        id_prefixes = [line.strip() for line in f]
    
    # Filter the FASTA file based on matching prefixes
    with open(output_file, "w") as out_f:
        for record in SeqIO.parse(fasta_file, "fasta"):
            if any(record.id.startswith(prefix) for prefix in id_prefixes):
                out_f.write(f">{record.id}\n{str(record.seq)}\n")

# Example usage
fasta_file = "/home/mkabir3/Research/27_Dispredict3_colab/inputs/input.fasta"
id_list_file = "shortList.txt"  # File containing ID prefixes, one per line
output_file = "filtered_output.fasta"

filter_fasta_by_partial_id(fasta_file, id_list_file, output_file)


# # Example usage
# fasta_file = "input.fasta"
# id_list_file = "shortList.txt"  # File containing IDs, one per line
# output_file = "filtered_output.fasta"

# filter_fasta_by_id(fasta_file, id_list_file, output_file)
```
