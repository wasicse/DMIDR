```python

from Bio import SeqIO
import re

count = 0
longproteinscount = 0
average_length = 0
newfile=open('./inputs/processedinput.fasta','w')
longproteins=open('./inputs/longproteins.fasta','w')
proteinLengths=open('./inputs/proteinLengths.txt','w')
mapfile=open('./inputs/mapfile.txt','w')
with open('./inputs/input.fasta', 'r') as f:
  for record in SeqIO.parse(f, 'fasta'):
    count += 1
    # print(record.id)
    # print(len(record.seq))
    if(len(record.seq) <= 5000):
      
      mapfile.write(record.id+ ', ')
      cleanrecordid=record.id.replace(r'\b', '')
      cleanrecordid = re.sub('[^0-9a-zA-Z]+', '', cleanrecordid)
      mapfile.write(cleanrecordid+ '\n')
      newfile.write('>' + cleanrecordid + '\n')
      newfile.write(str(record.seq) + '\n')
      average_length = average_length + len(record.seq)
      print(cleanrecordid+","+str(len(record.seq)),file=proteinLengths)

    else:
      # print('too long')
      longproteinscount=longproteinscount+1
      print(record.id,file=longproteins)
      print(len(record.seq),file=longproteins)
      
print("Total number of proteins", count)
print("Long proteins with length > 5000 : ",longproteinscount)
print("Feasible number of proteins: ",count-longproteinscount)
print("Average length: ",average_length/(count-longproteinscount))
newfile.close()
mapfile.close()
proteinLengths.close()
```

```python
split=(count-longproteinscount)/20
print(int(split))
```
